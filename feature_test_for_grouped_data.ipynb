### Analysis

from Clean_df, y was defined as a list of boolean data whether voter_turnout ratio is larger than or equal to 75% at each row.

```python
y = clean_df['Voter_Turnout'] >= 75
X = clean_df.copy()
del X['Voter_Turnout']
```

```python
X.head()
```

```python
feature_name = list(X.columns)
num_feats = 10
```

Feature extraction using Random Forest Classifier and SelectFromModel

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100), max_features=num_feats)
```

```python
embeded_rf_selector.fit(X, y)
```

```python
embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X.loc[:,embeded_rf_support].columns.tolist()
print(str(len(embeded_rf_feature)), 'selected features')
```

```python
embeded_rf_feature
```

Step to apply Random Forest Classifier on selected features set

```python
X_cleaned_feature = X[['E11B19013e1',
 'E12B19113e1',
 'E12B20002e1',
 'E12B20002e2',
 'E12B20002e3',
 'H18B25081e1',
 'D03B02001e7',
 'S07B15003e23-25',
 'E11B19001e11-12',
 'E11B19001e16-17']]
X_cleaned_feature.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_cleaned_feature,y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,max_depth=2)
rf = rf.fit(X_train,y_train)
```

```python
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```
